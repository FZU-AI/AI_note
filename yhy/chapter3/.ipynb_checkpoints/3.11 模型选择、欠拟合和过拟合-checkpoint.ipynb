## 3.11.1 训练误差和泛化误差
- 训练误差(training error)：数据集上表现的误差
- 泛化误差(generalization error)：测试集上表现的误差
- 机器学习模型应该关注降低泛化误差，我们要做的是能够准确预测出我们未曾见过的情况

## 3.11.2 模型选择(model selection)
- 以多层感知机为例，我们可以选择隐藏层的个数、各个隐藏层的隐藏单元和激活函数
- 模型选择时要使用验证数据集(validation data set)
    - 预留一部分在训练数据集和测试数据集以外的数据来进行模型选择，这部分数据叫做（验证集validation set）
- K折交叉验证(K-fold cross-validation)
    - 我们把原始训练数据集分割成K个不重合的子数据集，然后我们做K次模型训练和验证。每一次，我们使用一个子数据集验证模型，并使用其他K-1个子数据集来训练模型。最后，对K次训练误差和验证误差分别求平均
    ![1](../image/3.11_1.png)

## 3.11.3 欠拟合和过拟合
- 欠拟合(underfitting)：模型无法得到较低的训练误差
- 过拟合(overfitting)：模型的训练误差要远小于测试误差
- 导致两种拟合问题的因素
    - 模型复杂度:我们想要做的事情是，找到一个K阶多项式，使其结果能够尽可能的与标签y靠近，多项是函数拟合使用平方损失函数。一阶多项式函数拟合又叫线性拟合
        $$
        \hat{y} = b + \sum_{k=1}^K{x^kw_k}
        $$
        - 模型复杂度对欠拟合和过拟合的影响
        ![2](../image/3.11_2.png)
        模型复杂度过低，容易出现欠拟合，模型复杂度过高，容易出现过拟合
    - 训练数据集大小
        - 样本过少，模型参数过少时，过拟合更容易发生;**泛化误差不会随着数据集里样本数量增加而增大，换句话说，真金不怕火炼，一个好的模型，在经过大量测试后，表现应该是一个稳定值，不随测试样本数量的多少而波动**
    

## 3.11.4 多项式函数拟合实验

```python
# 导包
%matplotlib inline
import sys
sys.path.append("../")
import d2lzh as d2l
from mxnet import autograd, gluon, nd
from mxnet.gluon import data as gdata, loss as gloss, nn
```

### 生成数据集
- 使用如下三阶多项式函数来生成该样本的标签：
$$
    y = 1.2x - 3.4x^2 + 5.6x^3 + 5 + \epsilon
$$
- 噪声项$\epsilon$服从均值为0、标准差为0.1的正态分布。测试集和训练集样本都设为100

```python
n_train, n_test, true_w, true_b = 100, 100, [1.2, -3.4, 5.6], 5
features = nd.random.normal(shape=(n_train + n_test, 1))
poly_features = nd.concat(features, nd.power(features, 2), nd.power(features, 3))
labels = (true_w[0] * poly_features[:,0] + true_w[1] * poly_features[:,1] + true_w[2] * poly_features[:,2] + true_b)
labels += nd.random.normal(scale=0.1 ,shape=labels.shape)
```

```python
features[:2], poly_features[:2], labels[:2]
```

### 定义，训练和测试模型

```python
def semilogy(x_vals, y_vals, x_label, y_label, x2_vals=None, y2_vals=None, legend=None, figsize=(3.5, 2.5)):
    d2l.set_figsize(figsize)
    d2l.plt.xlabel(x_label)
    d2l.plt.ylabel(y_label)
    d2l.plt.semilogy(x_vals, y_vals)
    if x2_vals and y_vals:
        d2l.plt.semilogy(x2_vals, y2_vals, linestyle=':')
        d2l.plt.legend(legend)
    
```

```python
# 这里使用定义好的平方损失函数
num_epochs, loss = 100, gloss.L2Loss()

def fit_and_plot(train_features, test_features, train_labels, test_labels):
    net = nn.Sequential()
    net.add(nn.Dense(1))
    net.initialize()
    batch_size = min(10, train_labels.shape[0])
    train_iter = gdata.DataLoader(gdata.ArrayDataset(train_features, train_labels), batch_size, shuffle=True)
    trainer = gluon.Trainer(net.collect_params(), 'sgd', {'learning_rate': 0.01})
    train_ls, test_ls = [],[]
    for _ in range(num_epochs):
        for X,y in train_iter:
            with autograd.record():
                l = loss(net(X), y)
            l.backward()
            trainer.step(batch_size)
        train_ls.append(loss(net(train_features), train_labels).mean().asscalar())
        test_ls.append(loss(net(test_features), test_labels).mean().asscalar())
    print('final epoch: train loss', train_ls[-1], 'test loss', test_ls[-1])
    semilogy(range(1,num_epochs + 1), train_ls,  'epochs', 'loss',
             range(1, num_epochs + 1), test_ls, ['train', 'test'])
    print('weight', net[0].weight.data().asnumpy(), '\nbias:', net[0].bias.data().asnumpy())
```

```python
# 正常
fit_and_plot(poly_features[:n_train,:], poly_features[n_train:, :], labels[:n_train], labels[n_train:])
```

```python
# 欠拟合
fit_and_plot(features[:n_train,:], features[n_train:, :], labels[:n_train], labels[n_train:])
```

```python
# 样本不足，过拟合
fit_and_plot(poly_features[0:2,:], poly_features[n_train:, :], labels[0:2], labels[n_train:])
```

### 小结
- 由于无法从训练误差估计泛化误差，因此一味地降低训练误差并不意味着泛化误差一定会降低。机器需系模型应关注降低泛化误差
- 可以使用验证数据集来进行模型选择
- 欠拟合指无法得到较低的训练误差，过拟合指训练误差远低于泛化误差
- **应选择复杂度合适的模型并避免使用过少的训练样本**

### 练习
- 如果用一个三阶多项式模型来拟合一个线性模型生成的数据，可能会有什么问题？
    - 答：模型太过复杂了，会出现过拟合的问题
    
- 在本节提到的三阶多项式拟合问题里，有没有可能把100个样本的训练误差的期望降到0,为什么？
    - 答：没有可能。除非这1000个样本中只有小于等于4个点不共线，这种情况才会使得loss为0,因此3次多项式最多可以完全拟合4个不共线的点

```python
n_train, n_test, true_w, true_b = 100, 100, [1.2, -3.4, 5.6, 4.3, 3.6], 5
features = nd.random.normal(shape=(n_train + n_test, 1))
poly_features = nd.concat(features, nd.power(features, 2), nd.power(features, 3), nd.power(features, 4), nd.power(features, 5))
labels = (true_w[0] * poly_features[:,0] + true_w[1] * poly_features[:,1] + true_w[2] * poly_features[:,2] + true_w[3] * poly_features[:,3] + true_w[4] * poly_features[:,4] + true_b)
labels += nd.random.normal(scale=0.1 ,shape=labels.shape)
```

```python
fit_and_plot(poly_features[:n_train,:], poly_features[n_train:, :], labels[:n_train], labels[n_train:])
```
